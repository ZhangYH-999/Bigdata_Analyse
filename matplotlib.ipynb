### 1. plot

（折线图）

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'r', marker='o', linestyle='dotted')
plt.show()

# marker:      标记      o * . , x + p s 
# ms:          标记大小  marker size
# linestyle:   线类型    - solid  : dotted   -- dashed    -. dashed/dotted
# linewidth:   线宽
# color/c:     线颜色   'r'  '#4CAF50'

# 简写：
# plt.plot(ypoints, 'o:r')
```

### 2. scatter

（散点图）

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(low=1, high=11, size=50)
y = x + np.random.randint(1, 5, size=x.size)
data = np.column_stack((x, y))

plt.scatter(x=x, y=y,s=20, marker='o', c='r', edgecolor='b')
plt.title('Scatter: $x$ versus $y$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

# s=20               散点大小
# color='r'          散点颜色(也可以是一个列表，控制每个点的颜色)
# marker="o"         散点形状
# linewidths=3       边框的宽度
# edgecolors='r'     边框的颜色
# alpha=0.9          透明度
# label='线性'       图例
```

### 3. bar

（柱形图）

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

# 竖直
plt.bar(x, y, color = "green", width=0.3)
plt.show()

# color=['b']*10,        # 柱子的颜色
# width=0.5,             # 柱子宽度
# bottom=1,              # 纵坐标的起始大小
# align='edge'           # 横坐标和柱子的相对位置


# 水平
plt.barh(x, y,  height = 0.5)
plt.show()


# 堆叠柱形图
plt.figure(figsize=(6,4))               # 设置画布大小
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
width = 0.35       

plt.bar(labels, men_means, width, 
        label='Men')
plt.bar(labels, women_means, width, 
        bottom=men_means,                  # 堆叠在men_means上
        label='Women')

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.legend()
plt.show()

# 对比柱形图
plt.figure(figsize=(6,4))               # 设置画布大小
y1 = np.random.randint(1,10,10)
y2 = np.random.randint(1,10,10)
width = 0.3

x = ['N{:>03d}'.format(i+1) for i in range(10)]
plt.bar(x,y1, width=width, label='A')
plt.bar([i + width for i in range(len(x))], y2, width=0.3, label='B')
plt.xlabel('$x$',fontsize=14) 
plt.ylabel('$y$',fontsize=14)
plt.ylim(0, 12)

for i in range(len(y1)):
    plt.annotate(y1[i], (i-0.1, y1[i]+0.1))       # 设置A数据的标注
for j in range(len(y2)):
    plt.annotate(y2[j], (j+0.2, y2[j]+0.1))       # 设置B数据的标注
plt.legend()            # 添加图例
plt.show()
```

### 4. hist

（直方图）

```python
import matplotlib.pyplot as plt
import numpy as np

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

plt.hist(x,
         bins=50, 
         label=('x', 'y'))
plt.legend(loc=(0.65, 0.8))
plt.title('Frequencies of $x$ and $y$')
plt.show()

# bins=50 or 'auto'
# density=True,       默认为false，显示的是频数统计结果，为True则显示频率统计结果
# facecolor='b',      柱体的颜色
# alpha=0.8,          柱体的透明度
# log=False,          bool，默认False,即y坐标轴是否选择指数刻度
# stacked=False       bool，默认为False，是否为堆积状图
```

### 5. pie

（饼图）

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, autopct="%0.2f%%")
plt.show() 
 
# labels=["Apples", "Bananas","Dates"]        # 标签
# colors=['red','blue','green', 'yellow'],    # 颜色设置
# shadow=True,                                # 阴影显示
# labeldistance = 1.1,                        # 设置label距图的距离
# autopct="%.2f%%",                           # 设置百分比的显示  
# startangle=150,                             # 旋转角度
# explode = [0.1, 0, 0, 0]                    # 每部分凸显距离
```

### 6. stackplot

（堆叠图）

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(50) + 1950
y = np.random.randint(0, 10, size=(3, 50)) + np.arange(50)

plt.stackplot(x, y, labels=['Eastasia', 'Eurasia', 'Oceania'])
plt.title('Combined debt growth over time')
plt.ylabel('Total debt')
plt.legend()
plt.show()
```

### 7. heatmap
（热力图）

```python
plt.figure(figsize=(6,6))                           # 设置画布大小
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd", "Cornylee Corp"]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


plt.imshow(harvest, 
           cmap='autumn',    # 颜色选择 https://matplotlib.org/stable/tutorials/colors/colormaps.html
           aspect='equal',    # 用于指定热图的单元格的大小，默认值为'equal',此时单元格用于是一个方块，
                             # 当设置为'auto'时，会根据画布的大小动态调整单元格的大小
           alpha=0.8,        # 设置透明度
           origin='lower',   # orign参数指定绘制热图时的方向，默认值为upper,  此时热图的右上角为(0, 0), 
                             # 当设置为lower时，热图的左下角为(0,0)
           vmin=0, vmax=2    # vmin和vmax参数用于限定数值的范围，只将vmin和vmax之间的值进行映射
           )
plt.colorbar()           # 颜色条


# 更改坐标值
plt.xticks(np.arange(7),vegetables, rotation=90, fontsize=10)
plt.yticks(np.arange(7),farmers, rotation=20, fontsize=10)

# 每个方格上标记数值
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = plt.text(j, i, harvest[i, j],
                       ha="center", va="center", color="black", fontsize=8)

plt.title("heatmap")
# fig.tight_layout()
plt.show()
```

### 7. subplots

（子图划分）

**方法一：**

通过 `plt.subplots(1, 2)` 创建一个Figure对象和一个Axes对象的numpy数组，可用 `ax.flatten()` 展平数组

- `$` 包含的文本使用Tex标记

```python
x = np.random.randint(low=1, high=11, size=50)
y = x + np.random.randint(1, 5, size=x.size)
data = np.column_stack((x, y))

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2,
                               figsize=(8, 4))

ax1.scatter(x=x, y=y, marker='o', c='r', edgecolor='b')
ax1.set_title('Scatter: $x$ versus $y$')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')

ax2.hist(data, bins=np.arange(data.min(), data.max()),
         label=('x', 'y'))
ax2.legend(loc=(0.65, 0.8))
ax2.set_title('Frequencies of $x$ and $y$')
ax2.yaxis.tick_right()
```

```python
gridsize = (3, 2)
fig = plt.figure(figsize=(8, 6))
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=2, rowspan=2)
ax2 = plt.subplot2grid(gridsize, (2, 0))
ax3 = plt.subplot2grid(gridsize, (2, 1))
```
