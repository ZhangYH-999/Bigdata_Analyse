### Part 1

#### 1.1 数据处理

```python
import pandas as pd

ori_data = pd.read_csv('tmall_order_report.csv')
ori_data.head(5)
```

```python
ori_data.describe()
```

```python
ori_data.info()
```

```python
ori_data.columns = ori_data.columns.str.strip()  # 列名去空格
ori_data.columns
```

```python
# 判断是否存在重复行（可用drop_duplicated()去重）
print(ori_data[ori_data.duplicated()].count())

print(ori_data.isnull().sum())
```

#### 1.2 数据可视化

```python
from pyecharts import options as opts
from pyecharts.charts import Map, Bar, Line
from pyecharts.components import Table
from pyecharts.options import ComponentTitleOpts
from pyecharts.faker import Faker
```

```python
ori_data['订单创建时间'] = pd.to_datetime(ori_data['订单创建时间'])
ori_data['订单付款时间'] = pd.to_datetime(ori_data['订单付款时间'])


daliy_order = ori_data.groupby(ori_data['订单创建时间'].apply(lambda x: x.strftime("%Y-%m-%d"))).agg({'订单编号':'count'}).to_dict()['订单编号']

c = (
    Line()
    .add_xaxis(list(daliy_order.keys()))
    .add_yaxis('订单量', list(daliy_order.values()))
)
c.render_notebook()
```

```python
from matplotlib import pyplot as plt
plt.plot(list(daliy_order.keys()), list(daliy_order.values()))
plt.show()
```
